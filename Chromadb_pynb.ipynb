```python
import spacy

# Load the spaCy model
nlp = spacy.load("en_core_web_sm")

# Define the text
text = ""
with open("agri.txt", "r") as file:
    text = file.read()
text = text.replace('\n', ' ')
# Process the text with spaCy
doc = nlp(text)

# Extract sentences
sentences = [sent.text.strip() for sent in doc.sents]
```

```python
import chromadb

client = chromadb.Client()
collection = client.create_collection("sciq_supports")
```

```python
collection.add(
    ids=[str(i) for i in range(len(sentences))],  # IDs are just strings
    documents=sentences
)
```

```python
query = "which soil is Rich in iron and aluminum"
results = collection.query(
    query_texts=query,
    n_results=5)
```
